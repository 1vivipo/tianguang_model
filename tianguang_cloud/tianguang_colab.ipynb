## 4. 创建项目文件

```
import os

# 创建目录
os.makedirs('tianguang_cloud/data', exist_ok=True)
os.makedirs('tianguang_cloud/models', exist_ok=True)
os.makedirs('tianguang_cloud/scripts', exist_ok=True)

print('目录创建完成！')
```

## 5. 上传训练数据

运行下面的单元格，然后选择你的训练数据文件

```
from google.colab import files

print('请上传训练数据文件 (training_data.txt)')
uploaded = files.upload()

for fn in uploaded.keys():
    os.rename(fn, 'tianguang_cloud/data/training_data.txt')
    print(f'数据已保存: tianguang_cloud/data/training_data.txt')
```

## 6. 开始训练
